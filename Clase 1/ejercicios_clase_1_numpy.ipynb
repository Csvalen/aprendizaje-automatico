```python
import numpy as np

# Ejercicio 1
e1 = np.zeros(10)
print("Ejercicio 1:", e1)

# Ejercicio 2
e2 = np.arange(1, 21)
print("Ejercicio 2:", e2)

# Ejercicio 3
e3 = np.arange(2, 51, 2)
print("Ejercicio 3:", e3)

# Ejercicio 5
e5 = np.random.rand(4,4)
print("Ejercicio 5:\n", e5)

# Ejercicio 6
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Ejercicio 6:", a + b)

# Ejercicio 7
m1 = np.random.randint(1,11,(3,3))
m2 = np.random.randint(1,11,(3,3))
print("Ejercicio 7:\n", np.dot(m1, m2))

# Ejercicio 8
e8 = np.random.randint(1,101,15)
print("Ejercicio 8:", e8)
print("Max:", np.max(e8))
print("Min:", np.min(e8))
print("Prom:", np.mean(e8))

# Ejercicio 9
e9 = np.arange(1,26).reshape(5,5)
print("Ejercicio 9:\n", e9)
print("Diagonal:", np.diag(e9))

# Ejercicio 10
e10 = np.random.randint(1,21,(5,5))
print("Ejercicio 10:\n", e10)
print("Filas:", np.sum(e10, axis=1))
print("Columnas:", np.sum(e10, axis=0))
```
