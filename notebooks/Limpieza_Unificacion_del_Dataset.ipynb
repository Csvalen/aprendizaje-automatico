## Introducción

El presente proyecto tiene como objetivo desarrollar un modelo de Aprendizaje Automático orientado a la predicción del consumo mensual de gas envasado en Tierra del Fuego.

Para ello se utilizarán datasets oficiales relacionados con consumo energético, cantidad de usuarios empadronados y variables climáticas correspondientes a la provincia.

El dataset energético contiene registros mensuales históricos de consumo de gas envasado y cantidad de usuarios, mientras que el dataset climático incluye variables meteorológicas como temperatura, precipitaciones y días con nieve correspondientes a Río Grande y Ushuaia.

A partir de ambas fuentes se construirá un dataset integrado orientado al entrenamiento de modelos supervisados de regresión.

## Recuperación de datos faltantes mediante Open-Meteo

```python
import requests

def obtener_clima_mensual_openmeteo(lat, lon, fecha_inicio, fecha_fin):
    url = "https://archive-api.open-meteo.com/v1/archive"

    params = {
        "latitude": lat,
        "longitude": lon,
        "start_date": fecha_inicio,
        "end_date": fecha_fin,
        "daily": [
            "temperature_2m_max",
            "temperature_2m_min",
            "temperature_2m_mean",
            "precipitation_sum",
            "snowfall_sum"
        ],
        "timezone": "America/Argentina/Ushuaia"
    }

    r = requests.get(url, params=params)
    r.raise_for_status()
    data = r.json()

    df = pd.DataFrame(data["daily"])
    df["fecha"] = pd.to_datetime(df["time"])
    df["anio"] = df["fecha"].dt.year.astype(str)
    df["mes_num"] = df["fecha"].dt.month

    mensual = df.groupby(["anio", "mes_num"]).agg(
        temp_max=("temperature_2m_max", "mean"),
        temp_min=("temperature_2m_min", "mean"),
        temp_media=("temperature_2m_mean", "mean"),
        precipitacion=("precipitation_sum", "sum"),
        nieve=("snowfall_sum", lambda x: (x > 0).sum())
    ).reset_index()

    meses = {
        1: "ENERO", 2: "FEBRERO", 3: "MARZO", 4: "ABRIL",
        5: "MAYO", 6: "JUNIO", 7: "JULIO", 8: "AGOSTO",
        9: "SEPTIEMBRE", 10: "OCTUBRE", 11: "NOVIEMBRE", 12: "DICIEMBRE"
    }

    mensual["mes"] = mensual["mes_num"].map(meses)

    return mensual.round(2)
```

```python
# corregir mayo 2021 temp_min_rg
# ==========================================
# CORRECCIÓN DE TEMPERATURA MÍNIMA
# RÍO GRANDE - MAYO 2021
# Fuente: Open-Meteo
# ==========================================
rg_mayo_2021 = obtener_clima_mensual_openmeteo(
    lat=-53.7860,
    lon=-67.7000,
    fecha_inicio="2021-05-01",
    fecha_fin="2021-05-31"
)

valor_temp_min_rg = rg_mayo_2021.loc[0, "temp_min"]

clima_limpio.loc[
    (clima_limpio["anio"].astype(str) == "2021") &
    (clima_limpio["mes"].str.upper().str.strip() == "MAYO"),
    "temp_min_rg"
] = valor_temp_min_rg

clima_limpio.loc[
    (clima_limpio["anio"].astype(str) == "2021") &
    (clima_limpio["mes"].str.upper().str.strip() == "MAYO")
]
```

```python
# recuperar Ushuaia 2024
ush_2024 = obtener_clima_mensual_openmeteo(
    lat=-54.8019,
    lon=-68.3030,
    fecha_inicio="2024-01-01",
    fecha_fin="2024-12-31"
)

for _, fila in ush_2024.iterrows():
    mask = (
        (clima_limpio["anio"].astype(str) == "2024") &
        (clima_limpio["mes"].str.upper().str.strip() == fila["mes"])
    )

    clima_limpio.loc[mask, "temp_max_ush"] = fila["temp_max"]
    clima_limpio.loc[mask, "temp_min_ush"] = fila["temp_min"]
    clima_limpio.loc[mask, "temp_media_ush"] = fila["temp_media"]
    clima_limpio.loc[mask, "precipitacion_ush"] = fila["precipitacion"]

clima_limpio.loc[
    clima_limpio["anio"].astype(str) == "2024",
    ["anio", "mes", "temp_max_ush", "temp_min_ush", "temp_media_ush", "precipitacion_ush", "nieve_ush"]
]
```

```python
# Rellenamos faltantes usando promedio histórico del mismo mes

columnas_rellenar = [
    "consumo_gas_kg",
    "cantidad_usuarios"
]

for col in columnas_rellenar:
    dataset_final[col] = dataset_final.groupby("mes")[col].transform(
        lambda x: x.fillna(x.mean())
    )

# Verificamos nuevamente
dataset_final.isnull().sum()
```

```python
# Mostrar algunos registros donde nieve_rg es NaN
dataset_final[dataset_final["nieve_rg"].isnull()][["anio","mes","nieve_rg"]].head(20)
```

```python
# Reemplazamos valores faltantes de nieve_rg
# por 0 (sin días de nieve)

dataset_final["nieve_rg"] = dataset_final["nieve_rg"].fillna(0)

# Convertimos a entero
dataset_final["nieve_rg"] = dataset_final["nieve_rg"].astype(int)

# Verificamos nuevamente
dataset_final.isnull().sum()
```

```python
# Completar nieve_ush de 2024 usando mediana histórica mensual

for mes in dataset_final["mes"].unique():

    mediana_mes = dataset_final.loc[
        (dataset_final["anio"].astype(str) != "2024") &
        (dataset_final["mes"] == mes),
        "nieve_ush"
    ].median()

    dataset_final.loc[
        (dataset_final["anio"].astype(str) == "2024") &
        (dataset_final["mes"] == mes) &
        (dataset_final["nieve_ush"].isna()),
        "nieve_ush"
    ] = mediana_mes
```

```python
#Convertir los NaN históricos restantes a 0

dataset_final["nieve_ush"] = dataset_final["nieve_ush"].fillna(0)
dataset_final["nieve_ush"] = dataset_final["nieve_ush"].astype(int)
dataset_final.isnull().sum()
```

```python
# Renombramos la columna anio a año

dataset_final.rename(
    columns={"anio": "año"},
    inplace=True
)

# Verificamos
dataset_final.head()
```
