```python
import pandas as pd

df = pd.read_csv("usuarios_win_mac_lin.csv")
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Variables
X = df.drop("clase", axis=1)
Y = df["clase"]

# Dividir datos
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Modelo
modelo = LogisticRegression(max_iter=1000)

# Entrenar
modelo.fit(X_train, Y_train)

# Predicción
Y_pred = modelo.predict(X_test)

# Resultados
print("Precisión:", accuracy_score(Y_test, Y_pred))
print("\nReporte:\n", classification_report(Y_test, Y_pred))
```

El modelo logró una precisión del 67%, lo que indica que puede clasificar correctamente la mayoría de los casos. Algunas clases se predicen mejor que otras, pero en general el modelo funciona de manera aceptable.
