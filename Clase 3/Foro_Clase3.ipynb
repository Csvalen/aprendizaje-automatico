```python
import pandas as pd

df = pd.read_csv('/Automobile.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['weight'], df['mpg'])

plt.xlabel('Peso')
plt.ylabel('MPG (consumo)')
plt.title('Relación entre peso y consumo')

plt.grid()
plt.show()
```

En el gráfico de dispersión se observa que a medida que aumenta el peso del vehículo, el consumo en mpg disminuye. Esto indica que los autos más pesados consumen más combustible.

```python
plt.figure(figsize=(8,5))
plt.hist(df['mpg'], bins=10, edgecolor='black')

plt.title('Distribución del consumo (mpg)')
plt.xlabel('MPG')
plt.ylabel('Frecuencia')

plt.show()
```

El histograma permite observar cómo se distribuyen los valores de consumo de combustible, mostrando cuáles son los más frecuentes.

```python
import matplotlib.pyplot as plt

# Convertir origin de texto a número
df['origin_num'] = df['origin'].map({
    'usa': 1,
    'europe': 2,
    'japan': 3
})

plt.figure(figsize=(8,5))
plt.boxplot([df['cylinders'], df['origin_num']], tick_labels=['cylinders', 'origin'])
plt.title('Boxplot de variables')
plt.show()
```

El boxplot permite analizar la dispersión de los datos y detectar posibles valores atípicos en las variables.

```python
tabla = df.groupby('model_year')['origin'].count()

tabla.plot(kind='bar')

plt.title('Cantidad de autos por año')
plt.xlabel('Año')
plt.ylabel('Cantidad')

plt.show()
```

El gráfico de barras muestra la cantidad de automóviles por año, lo que permite comparar su distribución en el tiempo.

La visualización de datos es fundamental en el aprendizaje automático, ya que permite entender mejor la información, identificar patrones y facilitar la toma de decisiones antes de aplicar modelos.
